# Sorted Result (1976 ~_)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
for year in range(1976, 2026):
    response = requests.get(f'http://www.nfd.com.tw/house/year/{year}.htm')
    response.encoding = 'big5'
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table')
    if table:
        data = []
        for row in table.find_all('tr'):
            cells = row.find_all('td')
            row_data = [cell.text.strip() for cell in cells]
            data.append(row_data)     
    else:
        print('error')   
    df = pd.DataFrame(data[1:], columns=data[0])
    df.to_csv(f'data/{year}.csv')
```

```python
all_df = pd.DataFrame(columns=data[0])
for year in range(1976, 2026):
    temp_df = pd.read_csv(f'data/{year}.csv', index_col=0)
    all_df = pd.concat([all_df, temp_df], ignore_index=True)
```
